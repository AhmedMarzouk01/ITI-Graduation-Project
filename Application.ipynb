```python
import pyodbc
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk
from PIL import Image, ImageTk  # To handle PNG images in Tkinter

# Database connection settings
server = 'DESKTOP-ASSIJIM'
database = 'Final_Project'
connection_string = f'DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};Trusted_Connection=yes;'

# Function to connect to the database
def connect_db():
    try:
        return pyodbc.connect(connection_string)
    except Exception as e:
        messagebox.showerror("Error", f"Database connection failed: {e}")
        return None

# Function to handle login as instructor or student
def login(role):
    if role == "Instructor":
        show_instructor_page()
    elif role == "Student":
        show_student_page()

# Function to display the instructor page
def show_instructor_page():
    clear_window()
    # Title
    tk.Label(root, text="Instructor", font=("Arial", 24), bg="darkred", fg="White", padx=700, pady=30).grid(row=0, column=0, columnspan=2)

    # Add Student Form
    title_label = tk.Label(root, text="Add Student", font=("Arial", 16), bg="Lightblue", fg="Black", width=20, height=2, padx=10, pady=5)
    title_label.grid(row=2, column=0, columnspan=2, pady=10)

    tk.Label(root, text="First Name").grid(row=4, column=0)
    fname_entry = tk.Entry(root)
    fname_entry.grid(row=4, column=1)

    tk.Label(root, text="Last Name").grid(row=5, column=0)
    lname_entry = tk.Entry(root)
    lname_entry.grid(row=5, column=1)

    tk.Label(root, text="Birth Date (YYYY-MM-DD)").grid(row=6, column=0)
    birthdate_entry = tk.Entry(root)
    birthdate_entry.grid(row=6, column=1)

    tk.Label(root, text="City").grid(row=7, column=0)
    city_entry = tk.Entry(root)
    city_entry.grid(row=7, column=1)

    tk.Label(root, text="Email").grid(row=8, column=0)
    email_entry = tk.Entry(root)
    email_entry.grid(row=8, column=1)

    tk.Label(root, text="Password").grid(row=9, column=0)
    password_entry = tk.Entry(root, show='*')
    password_entry.grid(row=9, column=1)

    tk.Label(root, text="Gender").grid(row=10, column=0)
    gender_entry = tk.Entry(root)
    gender_entry.grid(row=10, column=1)

    tk.Label(root, text="Track ID").grid(row=11, column=0)
    track_id_entry = tk.Entry(root)
    track_id_entry.grid(row=11, column=1)

    tk.Button(root, text="Add Student", font=("Arial", 13), bg="darkred", fg="White", command=lambda: add_student(
        fname_entry, lname_entry, birthdate_entry, city_entry, email_entry, password_entry, gender_entry, track_id_entry
    )).grid(row=12, column=0, columnspan=2, pady=10)

    # Add Exam Form
    tk.Label(root, text="Add Exam", font=("Arial", 16), bg="Lightblue", fg="Black", width=20, height=2, padx=10, pady=5).grid(row=15, column=0, columnspan=2, pady=10)

    tk.Label(root, text="Exam Type").grid(row=16, column=0)
    exam_type_entry = tk.Entry(root)
    exam_type_entry.grid(row=16, column=1)

    tk.Label(root, text="Exam Date (YYYY-MM-DD)").grid(row=17, column=0)
    exam_date_entry = tk.Entry(root)
    exam_date_entry.grid(row=17, column=1)

    tk.Label(root, text="Start Time (HH:MM:SS)").grid(row=18, column=0)
    start_time_entry = tk.Entry(root)
    start_time_entry.grid(row=18, column=1)

    tk.Label(root, text="End Time (HH:MM:SS)").grid(row=19, column=0)
    end_time_entry = tk.Entry(root)
    end_time_entry.grid(row=19, column=1)

    tk.Label(root, text="Total Hours").grid(row=20, column=0)
    total_hours_entry = tk.Entry(root)
    total_hours_entry.grid(row=20, column=1)

    tk.Label(root, text="Year").grid(row=21, column=0)
    year_entry = tk.Entry(root)
    year_entry.grid(row=21, column=1)

    tk.Label(root, text="Total Grades").grid(row=22, column=0)
    total_grades_entry = tk.Entry(root)
    total_grades_entry.grid(row=22, column=1)

    tk.Label(root, text="Instructor ID").grid(row=23, column=0)
    instructor_id_entry = tk.Entry(root)
    instructor_id_entry.grid(row=23, column=1)

    tk.Button(root, text="Add Exam", font=("Arial", 13), bg="darkred", fg="White", command=lambda: add_exam(
        exam_type_entry, exam_date_entry, start_time_entry, end_time_entry, total_hours_entry, year_entry, total_grades_entry, instructor_id_entry
    )).grid(row=24, column=0, columnspan=2, pady=10)

    # Return to Login button
    tk.Button(root, text="Return to Login", font=("Arial", 13), bg="darkred", fg="White", command=show_login_page).grid(row=25, column=0, columnspan=2, pady=20)

# Function to display the student page
def show_student_page():
    clear_window()
    root.title("Student - View Courses, Exams, and Grades")

    tk.Label(root, text="Query Courses and Grades", font=("Arial", 24), bg="darkred", fg="White", padx=500, pady=50).grid(row=0, column=0, columnspan=3, pady=0)

    tk.Label(root, text="Student ID", font=("Arial", 22)).grid(row=4, column=0)
    student_id_entry = tk.Entry(root)
    student_id_entry.grid(row=4, column=1)

    tk.Button(root, text="Query Courses", font=("Arial", 13), bg="darkred", fg="White", command=lambda: query_courses(student_id_entry)).grid(row=4, column=2, padx=10)

    tk.Label(root, text="Exam ID", font=("Arial", 22)).grid(row=20, column=0)
    exam_id_entry = tk.Entry(root)
    exam_id_entry.grid(row=20, column=1)

    tk.Button(root, text="Query Exam", font=("Arial", 13), bg="darkred", fg="White", command=lambda: query_exam(exam_id_entry)).grid(row=20, column=2, padx=10)

    tk.Label(root, text="Student ID (For Grade)", font=("Arial", 22)).grid(row=50, column=0)
    student_id_grade_entry = tk.Entry(root)
    student_id_grade_entry.grid(row=50, column=1)

    tk.Button(root, text="Query Grade", font=("Arial", 13), bg="darkred", fg="White", command=lambda: query_grade(student_id_grade_entry)).grid(row=50, column=2, padx=10)

    # Return to Login button
    tk.Button(root, text="Return to Login", font=("Arial", 13), bg="darkred", fg="White", command=show_login_page).grid(row=60, column=0, columnspan=3, pady=20)

# Function to clear the current window's content
def clear_window():
    for widget in root.winfo_children():
        widget.destroy()

# Function to add student
def add_student(fname_entry, lname_entry, birthdate_entry, city_entry, email_entry, password_entry, gender_entry, track_id_entry):
    fname = fname_entry.get()
    lname = lname_entry.get()
    birthdate = birthdate_entry.get()
    city = city_entry.get()
    email = email_entry.get()
    password = password_entry.get()
    gender = gender_entry.get()
    track_id = track_id_entry.get()

    if all([fname, lname, birthdate, city, email, password, gender, track_id]):
        conn = connect_db()
        if conn:
            cursor = conn.cursor()
            try:
                cursor.execute("""
                    INSERT INTO Student (Student_Fname, Student_Lname, Student_BrithDate, 
                                        Student_City, Student_Email, Student_Password, Student_Gender, Track_Id)
                    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                """, fname, lname, birthdate, city, email, password, gender, track_id)
                conn.commit()
                messagebox.showinfo("Success", "Student added successfully.")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to add student: {e}")
            finally:
                conn.close()
    else:
        messagebox.showwarning("Input Error", "Please fill all the fields.")

# Function to add exam
def add_exam(exam_type_entry, exam_date_entry, start_time_entry, end_time_entry, total_hours_entry, year_entry, total_grades_entry, instructor_id_entry):
    exam_type = exam_type_entry.get()
    exam_date = exam_date_entry.get()
    start_time = start_time_entry.get()
    end_time = end_time_entry.get()
    total_hours = total_hours_entry.get()
    year = year_entry.get()
    total_grades = total_grades_entry.get()
    instructor_id = instructor_id_entry.get()

    if all([exam_type, exam_date, start_time, end_time, total_hours, year, total_grades, instructor_id]):
        conn = connect_db()
        if conn:
            cursor = conn.cursor()
            try:
                cursor.execute("""
                    INSERT INTO Exam (Exam_Type, Exam_Date, Start_Time, End_Time, 
                                      Total_Hours, Year, Total_Grades, Instructor_Id)
                    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
                """, exam_type, exam_date, start_time, end_time, total_hours, year, total_grades, instructor_id)
                conn.commit()
                messagebox.showinfo("Success", "Exam added successfully.")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to add exam: {e}")
            finally:
                conn.close()
    else:
        messagebox.showwarning("Input Error", "Please fill all the fields.")

# Function to query courses
def query_courses(student_id_entry):
    student_id = student_id_entry.get()
    if student_id:
        conn = connect_db()
        if conn:
            cursor = conn.cursor()
            try:
                cursor.execute("SELECT * FROM Courses WHERE Student_Id = ?", student_id)
                courses = cursor.fetchall()
                if courses:
                    courses_info = "\n".join([str(course) for course in courses])
                    messagebox.showinfo("Courses", courses_info)
                else:
                    messagebox.showinfo("Courses", "No courses found.")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to query courses: {e}")
            finally:
                conn.close()
    else:
        messagebox.showwarning("Input Error", "Please enter a Student ID.")

# Function to query exam
def query_exam(exam_id_entry):
    exam_id = exam_id_entry.get()
    if exam_id:
        conn = connect_db()
        if conn:
            cursor = conn.cursor()
            try:
                cursor.execute("SELECT * FROM Exam WHERE Exam_Id = ?", exam_id)
                exam = cursor.fetchone()
                if exam:
                    messagebox.showinfo("Exam", str(exam))
                else:
                    messagebox.showinfo("Exam", "Exam not found.")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to query exam: {e}")
            finally:
                conn.close()
    else:
        messagebox.showwarning("Input Error", "Please enter an Exam ID.")

# Function to query grade
def query_grade(student_id_grade_entry):
    student_id = student_id_grade_entry.get()
    if student_id:
        conn = connect_db()
        if conn:
            cursor = conn.cursor()
            try:
                cursor.execute("SELECT * FROM Grades WHERE Student_Id = ?", student_id)
                grades = cursor.fetchall()
                if grades:
                    grades_info = "\n".join([str(grade) for grade in grades])
                    messagebox.showinfo("Grades", grades_info)
                else:
                    messagebox.showinfo("Grades", "No grades found.")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to query grades: {e}")
            finally:
                conn.close()
    else:
        messagebox.showwarning("Input Error", "Please enter a Student ID.")

# Function to show the login page
def show_login_page():
    clear_window()
    root.title("Examination System Login")

    tk.Label(root, text="Welcome to Examination System", font=("Arial", 24), bg="darkred", fg="White", padx=500, pady=30).grid(row=0, column=0, columnspan=2)

    tk.Button(root, text="Instructor Login", font=("Arial", 16), command=lambda: login("Instructor"), bg="Lightblue", fg="Black").grid(row=1, column=0, padx=20, pady=20)
    tk.Button(root, text="Student Login", font=("Arial", 16), command=lambda: login("Student"), bg="Lightblue", fg="Black").grid(row=1, column=1, padx=20, pady=20)

# Main application window
root = tk.Tk()
root.geometry("800x600")
show_login_page()
root.mainloop()
```
